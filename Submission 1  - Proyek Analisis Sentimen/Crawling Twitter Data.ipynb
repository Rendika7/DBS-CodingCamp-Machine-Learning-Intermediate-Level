```python
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'www\.\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.encode('ascii', 'ignore').decode('ascii')
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()
    text = ' '.join([word for word in text.split() if len(word) > 1])

    return text
```

```python
Processed_additional_train['text_clean'] = Processed_additional_train['full_text'].apply(clean_text)
```

### Paired Lexicon

```python
# drop nilai nan di dalam corpus lexicon, yang kata formalnya tidak ada
lexicon_corpus = lexicon_corpus.dropna(subset=['formal'])
lexicon_corpus = lexicon_corpus.drop_duplicates(subset=['slang'])
```

```python
def replace_slang(text, slang_formal_pairs):
    for slang, formal in slang_formal_pairs:
        text = re.sub(r'\b{}\b'.format(re.escape(slang)), formal, text)
    return text

slang_formal_pairs = list(zip(lexicon_corpus['slang'], lexicon_corpus['formal']))
```

```python
# Fungsi untuk membersihkan teks
def clean_text(text, slang_formal_pairs = slang_formal_pairs):
    # Lowercasing
    text = text.lower()
    # Mengganti kata-kata dengan kata yang sesuai dari lexicon_corpus
    text = replace_slang(text, slang_formal_pairs)

    return text
```

```python
# Menerapkan fungsi ke kolom text
Processed_additional_train['text_clean_2'] = Processed_additional_train['text_clean'].apply(clean_text)
```

### Stemming & Stopword Removal

```python
# Download stopwords and wordnet
nltk.download(['stopwords', 'wordnet'])

# Initialize Sastrawi stemmer
factory = StemmerFactory()
stemmer_sastrawi = factory.create_stemmer()

# Define stop words from NLTK
stop_words_nltk = set(stopwords.words('indonesian'))

# Define additional stop words from Sastrawi or custom corpus
stop_words_sastrawi = set(stopword_corpus["Stopword"])

# Combine stop words
stop_words = stop_words_nltk.union(stop_words_sastrawi)

# Print the combined stop words
# stop_words
```

```python
# Define the preprocessing function
def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # Tokenize the text into words
    words = text.split()

    # Remove stopwords and apply stemming
    processed_words = []
    for word in words:
        if word.lower() not in stop_words:  # NLTK Stopword
            stemmed_word = stemmer_sastrawi.stem(word)  # Sastrawi Stemming
            processed_words.append(stemmed_word)

    # Join the processed words back into a single string
    return " ".join(processed_words)
```

```python
Processed_additional_train['text_clean_3'] = Processed_additional_train['text_clean_2'].astype(str).apply(preprocess_text)
```

### Applying Polarity Score

```python
lexicon_positive = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/9. DBS Foundation x Dicoding Kelas Menengah 2024/Additional Data/positive.csv')
lexicon_negative = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/9. DBS Foundation x Dicoding Kelas Menengah 2024/Additional Data/negative.csv')

lexicon_negative = lexicon_negative.rename(columns = {" word":"word"})
lexicon_positive = lexicon_positive.rename(columns = {" word":"word"})
```

```python
# Konversi lexicon ke dictionary
lexicon_positive_dict = dict(zip(lexicon_positive['word'], lexicon_positive['weight']))
lexicon_negative_dict = dict(zip(lexicon_negative['word'], lexicon_negative['weight']))
```

```python
def labeling(kalimat):
    score = 0
    words = kalimat.split()
    for word in words:
        if word in lexicon_positive_dict:
            score += lexicon_positive_dict[word]
        if word in lexicon_negative_dict:
            score += lexicon_negative_dict[word]

    if score > 0:
        polarity = 'positive'
    elif score == 0:
        polarity = 'neutral'
    else:
        polarity = 'negative'

    return score, polarity
```

```python
# Aplikasikan fungsi labeling pada kolom text
Processed_additional_train[['polarity_score', 'label based on polarity']] = Processed_additional_train['text_clean_3'].apply(lambda x: pd.Series(labeling(x)))
```

## 2) Transformers Model Labeling

model used :
- https://huggingface.co/ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa

- https://huggingface.co/mdhugol/indonesia-bert-sentiment-classification

```python
import transformers
print("Transformers version:", transformers.__version__)
from transformers import pipeline

sentiment_task  = pipeline(
    model="mdhugol/indonesia-bert-sentiment-classification",
    return_all_scores=True
)
```

```python
def predict_sentiment(df):
    # Create empty lists to store results
    id_sentiments = []
    id_scores = []

    # Iterate over each row in the dataframe
    for index, row in df.iterrows():
        # Predict sentiment for id_full_text
        id_result = sentiment_task(row['text_clean_3'])
        result = id_result[0]
        # Find the index of the highest score
        max_score_index = max(range(len(result)), key=lambda i: result[i]['score'])

        # Extract the label and highest score
        max_label = result[max_score_index]['label']
        max_score = result[max_score_index]['score']

        id_sentiments.append(max_label)
        id_scores.append(max_score)

    # Add sentiment and score columns to the dataframe
    df['label based on roBERTa Model'] = id_sentiments
    df['sentiment score'] = id_scores

    return df
```

```python
Processed_additional_train = predict_sentiment(Processed_additional_train)
```

```python
# Dictionary to replace labels
label_replace = {
    'LABEL_0': 'positive',
    'LABEL_1': 'neutral',
    'LABEL_2': 'negative'
}

# Replace labels in the DataFrame
Processed_additional_train['label based on roBERTa Model'] = Processed_additional_train['label based on roBERTa Model'].replace(label_replace)
```
